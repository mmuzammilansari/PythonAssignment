```python
# 1. Write a Python program to print the numbers from 1 to 10 using a for loop.
for i in range(1, 11):
    print(i)
```

```python
# 2. Write a Python program to print the numbers from 20 to 1 using a while loop.
i = 20
while i >= 1:
    print(i)
    i -= 1
```

```python
# 3. Write a program to print even numbers from 1 to 10.
for i in range(2, 11, 2):
    print(i)
```

```python
# 4. Write a program that prompts the user to enter a number n and prints all the numbers from 1 to n.
n = int(input("Enter a number: "))
for i in range(1, n + 1):
    print(i)
```

```python
# 5. Write a program that prompts the user to enter a number n, and then prints all the odd numbers between 1 and n.
n = int(input("Enter a number: "))
for i in range(1, n + 1, 2):
    print(i)
```

```python
# 6. Write a program that prints 'Happy Birthday!' five times on screen.
for _ in range(5):
    print("Happy Birthday!")
```

```python
# 7. Write a program that takes a number n as input and generates the first n terms of the series formed by squaring the natural numbers.
n = int(input("Enter a number: "))
print("The first", n, "terms of the series are:")
for i in range(1, n + 1):
    print(i ** 2, end=" ")
```

```python
# 8. Write a program that prompts the user to input a number and prints its multiplication table.
num = int(input("Enter a number: "))
for i in range(1, 11):
    print(f"{num} x {i} = {num * i}")
```

```python
# 9. Write a Python program to print the first 8 terms of an arithmetic progression starting with 3 and having a common difference of 4.
start = 3
difference = 4
for i in range(8):
    print(start + i * difference, end=" ")
```

```python
# 10. Write a Python program to print the first 6 terms of a geometric sequence starting with 2 and having a common ratio of 3.
start = 2
ratio = 3
term = start
for _ in range(6):
    print(term, end=" ")
    term *= ratio
```

```python
# 11. Write a program that asks the user for a positive integer value and calculates the sum of all integers up to that number.
n = int(input("Enter a positive integer: "))
print("The sum is:", sum(range(1, n + 1)))
```

```python
# 12. Write a program that calculates the sum of the reciprocals of all numbers from 1 to N.
n = int(input("Enter a positive integer: "))
reciprocal_sum = sum(1 / i for i in range(1, n + 1))
print(f"The sum of reciprocals from 1 to {n} is: {reciprocal_sum:.2f}")
```

```python
# 13. Write a program that prompts the user to enter a number and repeats this process 5 times, displaying the running total.
total = 0
for _ in range(5):
    num = int(input("Enter a number: "))
    total += num
print("The final running total is:", total)
```

```python
# 14. Write a program to calculate the factorial of a number entered by the user.
n = int(input("Enter a positive integer: "))
if n < 0:
    print("Factorial does not exist for negative numbers.")
elif n == 0:
    print("The factorial of 0 is 1.")
else:
    factorial = 1
    for i in range(1, n + 1):
        factorial *= i
    print(f"The factorial of {n} is {factorial}.")
```

```python
# 15. Write a program to calculate the power of a base to an exponent without using ** or math.pow().
base = int(input("Enter the base: "))
exponent = int(input("Enter the exponent: "))
result = 1
if exponent < 0:
    for _ in range(-exponent):
        result *= base
    result = 1 / result
else:
    for _ in range(exponent):
        result *= base
print(f"{base} to the power of {exponent} is {result}.")
```
