```python
# BMi Calculator and Classify
weight = float(input("Enter your weight (in kg): "))
height = float(input("Enter your height (in meters): "))
bmi = weight / (height ** 2)


if bmi < 18.5:
    print("Underweight")
elif bmi >= 18.5 and bmi <= 24.9:
    print("Normal weight")
elif bmi >= 25 and bmi <= 29.9:
    print("Overweight")
else:
    print("Obesity")
```

```python
#Calculate grade based on average marks
marks1 = float(input("Enter marks for subject 1: "))
marks2 = float(input("Enter marks for subject 2: "))
marks3 = float(input("Enter marks for subject 3: "))
average = (marks1 + marks2 + marks3) / 3

# Determine grade
if average >= 90:
    print("Grade: A")
elif average >= 80:
    print("Grade: B")
elif average >= 70:
    print("Grade: C")
elif average >= 60:
    print("Grade: D")
else:
    print("Grade: F")


```

```python
#  Roots of a quadratic equation
import math

a = float(input("Enter coefficient a: "))
b = float(input("Enter coefficient b: "))
c = float(input("Enter coefficient c: "))

discriminant = b**2 - 4*a*c

if discriminant > 0:
    root1 = (-b + math.sqrt(discriminant)) / (2 * a)
    root2 = (-b - math.sqrt(discriminant)) / (2 * a)
    print(f"Two real roots: {root1}, {root2}")
elif discriminant == 0:
    root = -b / (2 * a)
    print(f"Two equal roots: {root}")
else:
    real_part = -b / (2 * a)
    imaginary_part = math.sqrt(-discriminant) / (2 * a)
    print(f"Two complex roots: {real_part} + {imaginary_part}i, {real_part} - {imaginary_part}i")
```

```python
# Determine if a year is a leap year
year = int(input("Enter a year: "))
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print("Leap Year")
else:
    print("Not a Leap Year")
```

```python
# Calculate telephone bills
calls = int(input("Enter the number of calls: "))
bill = 200

if calls > 100:
    if calls <= 150:
        bill += (calls - 100) * 0.60
    elif calls <= 200:
        bill += 50 * 0.60 + (calls - 150) * 0.50
    else:
        bill += 50 * 0.60 + 50 * 0.50 + (calls - 200) * 0.40

print(f"Your telephone bill is Rs. {bill:.2f}")
```
